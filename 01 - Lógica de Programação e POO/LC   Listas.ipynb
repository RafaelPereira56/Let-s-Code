01 - Crie uma lista qualquer e faça um programa que imprima cada elemento da lista usando o for.

```python
listaTeste = [2, 5, 8, 25, 53, 0, 9]

for i in listaTeste:
  print(i) 

```

 02 Faça um programa que peça para o usuário digitar um número n e imprima uma lista com todos os números de 0 a n-1.

Exemplo: se o usuário digitar 5, o programa deve imprimir [0, 1, 2, 3, 4]

```python
listUser = int(input('Digite um número: '))

for i in range (listUser):

  print(i)
```

03 - Faça um programa que olhe todos os itens de uma lista e diga quantos deles são pares.

```python
listaTeste = [2, 5, 8, 25, 53, 0, 9, 12, 14, 16]

listaPar = []

for i in listaTeste:
  if i % 2 == 0:
    listaPar.append(i)

print(len(listaPar))    
```

04 - Faça um programa que imprima o maior número de uma lista, sem usar o método max().

```python
listaTeste = [-2, -5, -8, -25, -65, -9, -12, -14, -16]

ref = listaTeste [0]

for i in range (0, len(listaTeste)):
  if listaTeste [i] > ref:
    ref = listaTeste [i]

print(ref)
```

```python
numeros = []

numero = float(input('Digite um número: '))

while not numero == 0:
  if (len(numeros) == 0):
    numeros.insert(0,numero)
  else:
    cont = 0

    while cont < (len(numeros)):
      if numeros[cont] > numero: #perguntar o que esse colchetes faz
        numeros.insert(cont, numero)

        cont = len(numeros)

      cont += 1

      if (numeros[-1] != numero): #perguntar o que esse colchetes faz
        numeros.append(numero)

  while len(numeros) > 3:
    numeros.pop()

  numero = float(input('Digite um numero: '))

numeros.sort(reverse=True)

print(numeros) 
```

```python
def printMenu():
  print('1 - Depósito')
  print('2 - Retirada')
  print('3 - Saldo')
  print('4 - Sair do algoritmo')

def deposito(balanco):
  valor = float(input('Valor de deposito: '))

  return (balanco + valor)  

def retirada (balanco):
  valor = float(input('Valor do saque: '))

  if (balanco < valor):
    print ('Saldo insuficiente')

    return balanco
  else:
    return (balanco - valor)

def banco():
  opcao = -1

  balanco = 0

  while opcao != 4:
    printMenu()

    opcao = int(input())

    if opcao == 1:
      balanco = deposito(balanco)
    elif opcao == 2:
      balanco = retirada(balanco)
    elif opcao == 3:
      print('Saldo: R$ {}'.format(balanco))  

banco()
```

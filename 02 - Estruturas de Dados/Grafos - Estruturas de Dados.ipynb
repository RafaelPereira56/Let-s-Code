Implemente o grafo ponderado ilustrado pela figura abaixo, utilizando as representações baseadas em:

matriz de adjacências;
lista de adjacências.
Em seguida, exiba a matriz e a lista que representam o grafo.

```python
class Grafo:
    def __init__(self):
        self.adjacente_matriz = {}
        self.adjacente_lista = {}

    def add_vertice(self, indice):
        self.adjacente_matriz[indice] = {}
        self.adjacente_lista[indice] = []

    def add_aresta(self, indice_1, indice_2, peso):
        self.adjacente_matriz[indice_1][indice_2] = peso
        self.adjacente_matriz[indice_2][indice_1] = peso

        self.adjacente_lista[indice_1].append((indice_2, peso))
        self.adjacente_lista[indice_2].append((indice_1, peso))

    def mostrar_adjacente_matriz(self):
        representacao = '   |' 

        for i in self.adjacente_matriz:
            representacao += f' {i}'

        representacao += '\n'

        for i in self.adjacente_matriz:
            representacao += f' {i} |'

            for j in self.adjacente_matriz:
                representacao += f' {0 if not (self.adjacente_matriz.get(i) and self.adjacente_matriz.get(i).get(j)) else self.adjacente_matriz.get(i).get(j)}'

            representacao += '\n'

        return representacao

    def mostrar_adjacente_lista(self):
        representacao = ''

        for i in self.adjacente_lista:
            representacao += f'{i}: {self.adjacente_lista.get(i)}\n'

            return representacao

grafo = Grafo()

grafo.add_vertice(0)
grafo.add_vertice(1)
grafo.add_vertice(2)
grafo.add_vertice(3)
grafo.add_vertice(4)

grafo.add_aresta(0, 1, 3)
grafo.add_aresta(0, 3, 7)
grafo.add_aresta(0, 4, 8)
grafo.add_aresta(1, 2, 1)
grafo.add_aresta(1, 3, 4)
grafo.add_aresta(2, 3, 2)
grafo.add_aresta(3, 4, 3)

print(grafo.mostrar_adjacente_matriz())
print(grafo.mostrar_adjacente_lista())
```

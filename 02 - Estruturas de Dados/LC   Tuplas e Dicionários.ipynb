Crie uma função que recebe uma lista e retorna True se todos os seus elementos forem numéricos (int, float ou string contendo um int ou float) ou False do contrário. A função deve também retornar a lista tratada: transformar todas as entradas não numéricas em numéricas ou, no melhor caso, devolver a lista apenas.

```python
def is_numeric(element):
    if isinstance(element, int):
        return True, element

    if isinstance(element, float):
        return True, element
    
    if isinstance(element, str):
        if element.isnumeric:
            return True, float(element) if '.' in element else int(element)
        
        return False, None
    
    return False, element

def run(elements):
    formated_list = []

    for element in elements:
        numerical, element_as_numeral = is_numeric(element)

        if not numerical:
            return False, []
        
        formated_list.append(element_as_numeral)
    
    return True, formated_list

elements = [1, 3, '5.8', '123']

print(run(elements))
```

11 - Enunciado
Faça uma função que receba uma lista e conte quantas vezes cada elemento apareceu nessa lista. Essa função deverá guardar os dados em um dicionário no qual as chaves são os elementos da lista e os valores são a contagem de quantas vezes esse elemento aparece.

```python
def countList (aLista):
    dictionary = {}

    for element in aLista:
      if dictionary.get(element):
          dictionary[element] += 1
      else:
          dictionary[element] = 1

    print(dictionary)

countList([1, 2, 3, 4, 1, 2, 3])
```

12 - Faça um programa que fique pedindo uma resposta do usuário, entre 1, 2 e 3. Se o usuário digitar 1, o programa deve cadastrar um novo usuário nos moldes do exercício 10 e guardar esse cadastro num dicionário cuja chave será o CPF da pessoa. Quando o usuário digitar 2, o programa deve imprimir os usuários cadastrados; e se o usuário digitar 3, o programa deve fechar.

Exemplo do dicionário:

‘987.654.321-00’: {‘nome’: Maria, ‘idade’: 20, ‘email’: maria@mail.com}

```python
def cadastro():
    print('[1] - Add user')
    print('[2] - Show user')
    print('[3] - Exit')

    option = int(input('Enter the option: '))

    theDictionary = {}

    while option != 3:
        if option == 1:
            addUser = input('Write your CPF: ')

            name = input('Enter your name: ')

            age = input('Enter your age: ')

            email = input('Enter your email')

            if not theDictionary.get(addUser):
              theDictionary[addUser] = {
                    'name': name,
                    'age': age,
                    'email': email,               
                }
            else:
                print('Usuario ja cadastrado com esse CPF')

        elif option == 2:
                print(theDictionary)

        option = int(input('Entre com a opcao: '))   

cadastro()                     
```

Implemente um sistema de busca para o programa do exercício anterior. Isto é, se o usuário digitar 4, procure um determinado usuário pelo seu CPF.

```python
def cadastro():
    print('[1] - Add user')
    print('[2] - Show user')
    print('[3] - Exit')
    print('[4] - Look for user')

    option = int(input('Enter the option: '))

    theDictionary = {}

    while option != 3:
        if option == 1:
            addUser = input('Write your CPF: ')

            name = input('Enter your name: ')

            age = input('Enter your age: ')

            email = input('Enter your email')

            if not theDictionary.get(addUser):
              theDictionary[addUser] = {
                    'name': name,
                    'age': age,
                    'email': email,               
                }
            else:
                print('User already exist')

        elif option == 2:
                print(theDictionary)

        elif option == 4:
          addUser = input('Enter your CPF: ')   
            
        if theDictionary.get(addUser):
                print(theDictionary[addUser])
        else:
                print('User does not exist with this CPF')

        option = int(input('Enter the options: '))

cadastro()    
```
